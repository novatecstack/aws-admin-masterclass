# Histogram
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('players_21.csv')
```

```python
len(players)
```

```python
# Creating a Hist Plot
fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Number of Players')
ax1.set_xlabel('Overall score (bins)')

ax1.hist(x=players['overall'])

plt.show()
```

```python
# Changing bin size to 20 

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Number of Players')
ax1.set_xlabel('Overall score (bins)')

ax1.hist(x=players['overall'], bins=20)

plt.show()
```

```python
# Changing edgecolor of rectangles to black

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Number of Players')
ax1.set_xlabel('Overall score (bins)')

ax1.hist(x=players['overall'], bins=[0,25,50,75,100],edgecolor='black')

plt.show()
```

```python
# Adding xticks to align with bins

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Number of Players')
ax1.set_xlabel('Overall score (bins)')

plt.xticks([0,25,50,75,100],[0,25,50,75,100])

ax1.hist(x=players['overall'], bins=[0,25,50,75,100],edgecolor='black')

plt.show()
```

```python
# Horziontal alignment

fig = plt.figure(figsize = (5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('Number of Players')
ax1.set_ylabel('Overall score (bins)')

plt.yticks([0,25,50,75,100],[0,25,50,75,100])

ax1.hist(players['overall'], bins=[0,25,50,75,100], edgecolor='black',orientation='horizontal')

plt.show()
```
